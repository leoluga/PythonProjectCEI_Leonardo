```python
import struct

# Constants for field lengths in bytes
PRIMARY_HEADER_LENGTH = 6
SECONDARY_HEADER_LENGTH = 8  # Assuming GPS time is always present in TM SPs

# Define a function to parse the primary header
def parse_primary_header(data):
    primary_header_format = ">HBBH"
    primary_header = struct.unpack(primary_header_format, data[:PRIMARY_HEADER_LENGTH])
    
    version_number = (primary_header[0] >> 13) & 0b111
    packet_type = (primary_header[0] >> 12) & 0b1
    secondary_header_flag = (primary_header[0] >> 11) & 0b1
    apid = primary_header[0] & 0x07FF
    sequence_flags = (primary_header[1] >> 14) & 0b11
    sequence_count = primary_header[1] & 0x3FFF
    packet_data_length = primary_header[2]  # This is the length minus one

    return {
        "version_number": version_number,
        "packet_type": packet_type,
        "secondary_header_flag": secondary_header_flag,
        "apid": apid,
        "sequence_flags": sequence_flags,
        "sequence_count": sequence_count,
        "packet_data_length": packet_data_length
    }

# Define a function to parse the secondary header (GPS time)
def parse_secondary_header(data):
    secondary_header_format = ">II"
    secondary_header = struct.unpack(secondary_header_format, data[:SECONDARY_HEADER_LENGTH])
    week = secondary_header[0]
    ms = secondary_header[1]
    return {
        "week": week,
        "milliseconds": ms
    }

# Define a function to parse the packet data field
def parse_packet_data_field(data, secondary_header_flag):
    offset = PRIMARY_HEADER_LENGTH
    if secondary_header_flag:
        secondary_header = parse_secondary_header(data[offset:])
        offset += SECONDARY_HEADER_LENGTH
    else:
        secondary_header = None

    user_data = data[offset:-2]  # Exclude checksum
    checksum = struct.unpack(">H", data[-2:])[0]

    return {
        "secondary_header": secondary_header,
        "user_data": user_data,
        "checksum": checksum
    }

# Define a function to parse a complete CCSDS Space Packet
def parse_space_packet(data):
    primary_header = parse_primary_header(data)
    packet_data = parse_packet_data_field(data, primary_header["secondary_header_flag"])
    return {
        "primary_header": primary_header,
        "packet_data": packet_data
    }

# Read a binary file containing CCSDS Space Packets
file_path = 'decoded_satcs_dump\\adcs_housekeeping_only.out'
with open(file_path, 'rb') as f:
    data = f.read()

# Parse the first packet (assuming the packet length is known or can be determined)
space_packet = parse_space_packet(data)
print(space_packet)
```
